```python
def dist(a,b,ax=1):
    return np.linalg.norm(a-b,axis=ax)
```

```python
k=3
cx=np.random.randint(0,np.max(x)-20,size=k)
cy=np.random.randint(0,np.max(x)-20,size=k) # Returns 3 random points
cxy=np.array(list(zip(cx,cy)))
plt.scatter(f1,f2,s=8,color='g')
plt.scatter(cx,cy,s=50,color='r',marker='D')
```
