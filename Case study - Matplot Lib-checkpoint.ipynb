```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as mt
import seaborn as sns
%matplotlib inline

def digit(interval):
    if len(interval)==3:
        return 10
    elif len(interval)==4:
        return 100
    elif len(interval)==5:
        return 1000
    else:
        return 1

df_Churn = pd.read_csv("customer_churn.csv")
set1 =list(set(df_Churn['Contract']))
values=[]
for i in set1:
    filter1 = (df_Churn['Contract'] == i)
    df_new = df_Churn[filter1]
    count = int(df_new['Contract'].count())
    values.append(count)
plt.bar(set1,values,color='r',alpha=0.5)
plt.xlabel("Type of Contract")
plt.ylabel("count")
plt.title("Distribution of Contract")
plt.show()

min1=mt.floor(df_Churn['MonthlyCharges'].min())
max1=mt.ceil(df_Churn['MonthlyCharges'].max())
interval = round((max1-min1))
interval1 =str(interval)
digit = digit(interval1)

 
min2=round((min1-digit)/digit)*digit
max2=round((max1+digit)/digit)*digit

#print(digit,min1,max1,interval)
#print(digit,min2,max2,interval)
fig = plt.figure()
axes=plt.gca()
axes.set_xticks(np.arange(min2,max2,digit))
bins1=list(range(min2,max2,digit))
plt.hist(df_Churn['MonthlyCharges'],bins=bins1,color='cyan')
plt.show()


```

```python
x=list(df_Churn['TotalCharges'])
y=list(df_Churn['tenure'])
plt.scatter(x,y,marker ='D')
plt.show()
```

```python
set1 = list(set(df_Churn['PaymentMethod']))
print(set1)
df_Churn.boxplot(column=set1,by=list(df_Churn['MonthlyCharges']),showmeans=True)
plt.grid(True)
plt.show()
```
