```python
RFM_norml = RFM.drop("CustomerID",axis=1)
from sklearn.preprocessing import StandardScaler

SS=StandardScaler()
SS.fit_transform(RFM_norml)
```

```python
RFM_norml = pd.DataFrame(RFM_norml)
RFM_norml.columns = ['InvoiceNo','Spend','Recency']
RFM_norml.head()
```

```python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan

def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) # heuristic from article [1]
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

```python
hopkins(RFM_norml)
```

```python
from sklearn.cluster import KMeans
import seaborn as sns
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
K_clus=KMeans(n_clusters=3, init='random', n_init=10, max_iter=50)
K_clus.fit(RFM_norml)
labels=K_clus.predict(RFM_norml)
print(K_clus)
print("labels")
print(labels.shape)
centroids=K_clus.cluster_centers_
print(centroids)
```

```python
plt.scatter(RFM_norml['InvoiceNo'],RFM_norml['Spend'],c=labels,s=10)
plt.scatter(centroids[:,0],centroids[:,1],c='red',marker='*',s=100)
```

```python
plt.scatter(RFM_norml['InvoiceNo'],RFM_norml['Recency'],c=labels,s=10)
plt.scatter(centroids[:,0],centroids[:,2],c='red',marker='*',s=100)
```

```python
plt.scatter(RFM_norml['Spend'],RFM_norml['Recency'],c=labels,s=10)
plt.scatter(centroids[:,1],centroids[:,2],c='red',marker='*',s=100)
```

```python
n_clusters: The number of clusters to form as well as the number of centroids to generate
init: Method for initialization of the cluster centres
n_init: Number of time the k-means algorithm will be run with different centroid seeds. 
    The final results will be the best output of n_init consecutive runs in terms of inertia (explained below).
max_iter: Maximum number of iterations of the k-means algorithm for a single run in case it doesn't converge on its own
```

```python
from sklearn.metrics import silhouette_score
sse_ = []
for k in range(2, 15):
    kmeans = KMeans(n_clusters=k).fit(RFM_norml)
    sse_.append([k, silhouette_score(RFM_norml, kmeans.labels_)])
```

```python
plt.plot(pd.DataFrame(sse_)[0], pd.DataFrame(sse_)[1]);
```

```python
ssd = []
for num_clusters in list(range(1,21)):
    model_clus = KMeans(n_clusters = num_clusters, max_iter=50)
    model_clus.fit(RFM_norml)
    ssd.append(model_clus.inertia_)

plt.plot(ssd)
```
