```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0,100,10)
y = x * 2
fig =plt.figure(figsize=(10,5))
plt.plot(x,y, linewidth = 2.0, linestyle= ':', color='r',alpha=1,marker='x')
plt.title("my first graph")
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.legend(['Line1'])
plt.grid(True)
```

```python
x=np.arange(0,10)
y1=x*2
y2=x*3

plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title("Graph1")

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title("Graph2")

```

```python
x=np.arange(0,10)
y1=x*2
y2=x*3

plt.plot(x,y1,label ='Line1',color='r')
plt.plot(x,y2,label ='Line2',color='g')
plt.legend()
```

```python


x = np.arange(0,100,10)
y = x * 2
z= x * 1.5
fig =plt.figure(figsize=(10,5))
plt.stackplot(x,y, linewidth = 2.0, linestyle= ':', color='r',alpha=0.3)
plt.plot(x,z,linewidth=3.0,color='g',alpha=1)
plt.title("Area and line graph")
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.show()
```
