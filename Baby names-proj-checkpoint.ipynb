```python
year=[]
for i in range(1880,2021):
    year.append(pd.read_csv(f"yob{i}.txt",names=['name','gender','babies']))
    year[-1]['year']=i
```

```python
df_year=pd.concat(year)
```

```python
df_year
```

```python
df_piv=pd.pivot_table(df_year,index='year',columns='gender',values='babies',aggfunc='sum')
```

```python
df_piv
```

```python
#plt.figure(figsize=(50,30))
df_piv.plot(kind='barh',figsize=(10,40),grid=True)
```

```python
#pd.options.display.max_rows=100
sort_baby_names=df_year.sort_values(by='babies',ascending=False).reset_index(drop=True)
```
