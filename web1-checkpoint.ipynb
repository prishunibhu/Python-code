```python
import requests as rq
import pandas as pd
from bs4 import BeautifulSoup
```

```python
link='https://en.wikipedia.org/wiki/East_Asia'
link_data=rq.get(link).text
```

```python
print(link_data)
```

```python
soup=BeautifulSoup(link_data,'lxml')
print(soup.prettify())
```

```python
soup.head.text
```

```python
all_mv=soup.find_all('a',class_='mw-redirect')
print(all_mv)
```

```python
list1=[]
title1=[]
for link in all_mv:
    list1.append(link.get('href'))
    title1.append(link.get('title'))
```

```python
print(len(list1))
print(len(title1))
```

```python
df_list=pd.DataFrame(list1,columns=['links'])
df_list['title']=title1
df_list
```
