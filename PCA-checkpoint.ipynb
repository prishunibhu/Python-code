```python
def set_dummies(df_final,attribute):
    dummy = pd.get_dummies(df_final[attribute],prefix=attribute,drop_first=True)
    df_final_dummy=pd.concat([df_final,dummy],axis=1)
    df_final_dummy = df_final_dummy.drop(attribute,1)
    return df_final_dummy

df_final=set_dummies(df_final,'StreamingMovies')
```

```python
def draw_roc(actual,pred):
    fpr,tpr,thresholds=roc_curve(actual,pred,drop_intermediate=False)
    auc_score=roc_auc_score(actual,pred)
    plt.figure(figsize=(6,6))
    plt.plot(fpr,tpr,label='Roc curve(area=%0.2f)' %auc_score)
    plt.plot([0,1],[0,1],'k--')
    plt.xlim([0.0,1.0])
    plt.xlim([0.0,1.05])
    plt.xlabel('False positive rate or True negative rate')
    plt.ylabel('True positive rate')
    plt.legend()
    plt.show()
    
    return fpr,tpr,thresholds
```
