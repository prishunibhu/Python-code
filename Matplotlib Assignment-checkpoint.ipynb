```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df_Cars=pd.read_csv("Cars.csv")
fig =plt.figure(figsize=(100,50))
plt.plot(list(df_Cars['model']),list(df_Cars['hp']),color='r')
plt.xlabel("Model of the car",fontsize=50)
plt.ylabel("Horse power of the car",fontsize=50)
plt.title("Model vs Horse Power",fontsize=50)
plt.xticks(fontsize=10)
plt.yticks(fontsize=50)
plt.show()
```

```python
#assignment 2
df_Cars.info()
plt.bar(list(df_Cars['carb']),list(df_Cars['gear']),color='b')
plt.show()
df_Cars.describe()
```

```python
import math
plt.close('all')
bin_interval=math.ceil(df_Cars['wt'].max() - df_Cars['wt'].min())/30
bins1=np.arange(df_Cars['wt'].min(),df_Cars['wt'].max(),bin_interval)
fig =plt.figure(figsize=(100,50))
axes=plt.gca()
axes.set_xticks(bins1)
plt.xticks(fontsize=30)
plt.hist(list(df_Cars['wt']),bins1,color='blue',alpha=0.3)
plt.show()
```

```python
#Assignment 4

#plt.pie(Vallues1,labels=label1,startangle=60,autopct='%1.2f%%')
plt.close('all')
fig =plt.figure(figsize=(15,15))
axes=plt.gca()
plt.pie(list(df_Cars['cyl']),labels=list(df_Cars['model']),autopct='%1.1f%%')
plt.show()
```
