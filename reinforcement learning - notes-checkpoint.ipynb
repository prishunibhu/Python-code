```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Actions:
    def __init__(self,m):
        self.m=m
        self.N=0
        self.mean=0  # for epsilon greedy
        #self.mean = upper limit #for optimal method passed as the argument
    def choose(self):
        return np.random.randn() + self.m
    def update(self,x):
        self.N += 1
        self.mean = (1- 1.0/self.N)*self.mean + 1.0 /self.N *x
```

```python
def experiment(m1,m2,m3,esp,N):
    actions=[Actions(m1),Actions(m2),Actions(m3)]
    data=np.empty(N)
    for i in range(0,N):
        p=np.random.random()  # generates a number between 0 and 1
        print(p,"p")
        if p < esp:
            j=np.random.choice(3) 
        else:
            j=np.argmax([a.mean for a in actions]) #argmax function returns the index of the max value present.
        x=actions[j].choose()
        actions[j].update(x)
        data[i]=x
        print("data",i,data[i])
        
    cum_avg = np.cumsum(data) / (np.arange(N) + 1) 
    print("cum avg", i," ", cum_avg)
    plt.plot(cum_avg)
    plt.plot(np.ones(N) * m1)
    plt.plot(np.ones(N) * m2)
    plt.plot(np.ones(N) * m3)
    plt.xscale('log')
    plt.show()

    for a in actions:
        print("a.mean",a.mean)
    return cum_avg

            
```

```python
c_05=experiment(1.0,2.0,3.0,0.1,20)
```
