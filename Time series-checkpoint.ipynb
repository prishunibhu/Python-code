```python
Two types

Stationary - does not change with time. follows same pattern basis the mean,variance . 
nean/variance/covariance is same accross time
white noise is identified to be stationary
random walk is stationary - a series where current observartion is equal to previous + a random change.

The probability distribution of the time series is invariant is time. 


Non stationary - changes with time. mean/variance/co variance will be changed

4 different variations - trend,seasonal variation,cyclic variation,irregular variation or noise.

Missing values in time series:
    
*cannot take mean or median of entire data as it may not appropriate to fit into the trend of missing data - Mean Imputation

Many methods

* use the latest one before missing and copy to missing - Last carry forward
* take mean of before and after missing and fill missing - Linear interpolation
* seasonality + linear interpolation - if mar X iss missing, take mean of mar x-1 and mar x+1 to account seasonality.


Forecast models:
    

Naive method
Forecast = Last month’s sales

 

Simple average method
Forecast = Average of all past months’ sales

simple moving average

Average of past 12/6 months forecasts the future

Weighted moving average/simple exp smoothing:

Assigning weights to the rolling window. 

recent months= more weights
previous months = less weights

y(t+1) = a* t + b*(t-1) + c*(t-2)  /(a+b+c)

Testing for Stationary:
    
KPSS test is a unit root test(if a time series can be explained by a  trend line)

Null hypothesis : Series is stationary p > 0.05
Alternative hypothesis : series is non stationary p <=0.0.5
    
ADF test:

H0: Series is not stationary , p> 0.05
H1: Series is stationary, p1 <= 0.05

Changing non stationary to stationary:
    
Two steps:
    
* differrencing - simply take difference between two time plots and create a new plot. repeat the differencing to higher order
                  until it is observed to be stationar
    
Box cox plot = Choose a lamda value between -5 to 5 such that the forumla when applied converts my time series to 
                stationary.
    
y(lamda) =lamda^ y - 1 / lamda if lamda != 0
            =log(y), if lamda =0

Autocorellation:

To know if a time plot is co related with the time previous.

covariance[1,5] = covariance[2,6]

The co variance between two different time intervals is same.(remember - it is stationary process)

Two methods:
    
*Auto correlation - change is co relation in the past - how it effects in future.
*Partial autocorrelation

Test of goodness of TS model:
    
    Residuals of E(k) should have a white noise process. the e(k)(errors) should be un correlated

Now from the converted stationary model, we forecast and we have to convert to non stationary

* AR ,*MA,*ARMA,ARIMA,SARIMA,SARIMAX
Auto regression - Linear regression of past time series is used to predict future. Lags of high correlation is selected 
                    to build the linear regression equation

Moving average - pedicts future forecast using linear regression of past forecast errors

yhat = overall mean + errot(t-1) * theta.

q parameter is the window parameter to determine this which will Select q as the highest lag beyond which autocorrelation dies down

for eg , if q = 2

yhat = overall mean + errot(t-1) * theta + errot(t-2) * theta.


ARMA:

AR + MA - Builds a linear relationship on AR + MA

ARIMA: Auto Regressor Integrated moving average
    
Differencing + AR + MA

p,q,d(order of difference)

SARIMA: 
    
p,q,d and P Q D
Non-seasonal elements
p: Trend autoregression order
d: Trend difference order
q: Trend moving average order
Seasonal elements
m: The number of time steps for a single seasonal period
P: Seasonal autoregressive order
D: Seasonal difference order
Q: Seasonal moving average order
    
SARIMAX = SARIMA + exogeneous variable(that promotes non periodic peaks due to external factors. eg. a discount applied promotes sales)

Which method is used when?

*have less historical data(<10) -> noise -> simple moving average. eg - short term stock price data.
*recently opened store (10 months) - no noise - no seasons/no trend observed- naive method
*good historical data- no trend/no seasonal - Simple exponential smooting
```

# naive bayes

```python
train_len=120
df_train=df_air[0:train_len]
df_test=df_air[train_len:]

yhat_naive=df_test.copy()
yhat_naive['forecast'] = df_train['Pass_interpolate'][train_len-1]
```

```python
plt.figure(figsize=(12,4))
plt.plot(df_train['Pass_interpolate'],label='Train')
plt.plot(df_test['Pass_interpolate'],label='Test')
plt.plot(yhat_naive['forecast'],label='Forecast')
plt.legend()
plt.title("Naive method")
```

```python
import math
from sklearn.metrics import *
rmse_naive=math.sqrt(mean_squared_error(df_test['Pass_interpolate'],yhat_naive['forecast']))
print(rmse_naive)
mape = np.mean(np.abs(df_test['Pass_interpolate']-yhat_naive['forecast'])/df_test['Pass_interpolate'])* 100
print(mape)
results=pd.DataFrame({'Method': 'Naive','RMSE': [rmse_naive], 'MAPE':[mape]})
```
