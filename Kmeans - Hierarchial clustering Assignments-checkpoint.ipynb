```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

# Case study 2


```python
df_cars = pd.read_csv("cars.csv")
```

```python
df_cars.info()
```

```python
cars_features=df_cars[['hp','disp','mpg']]
arr1=np.array(cars_features)
print(arr1)
```

```python
plt.style.use("ggplot")
```

```python
kmeans_cars=KMeans(n_clusters=3)
kmeans_car=kmeans_cars.fit(arr1)
labels=kmeans_car.predict(arr1)
print("labels",labels)
centroids=kmeans_cars.cluster_centers_
print(centroids)
```

```python
xs=df_cars['mpg']
ys=df_cars['disp']
plt.scatter(xs,ys,c=labels,s=30)
centroids_x=centroids[:,2]
centroids_y=centroids[:,0]
plt.scatter(centroids_x,centroids_y,marker='D',s=100)
plt.show()
```

```python
plt.style.use("ggplot")
```

```python
plt.scatter(df_cust['Age'],df_cust['Spending Score (1-100)'],color='red')
```

```python
arr_kmeans=np.array(df_cust[['Age','Spending Score (1-100)']])
```

```python
#Elbow method to determine number of clusters
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters= i)
    kmeans.fit(arr_kmeans)
    wcss.append(kmeans.inertia_)

    #inertia_ is the formula used to segregate the data points into clusters
    
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans_cust=KMeans(n_clusters=5)
kmeans_cust=kmeans_cust.fit(arr_kmeans)
labels =kmeans_cust.predict(arr_kmeans)
centroids=kmeans_cust.cluster_centers_
print(centroids)
```

```python
plt.scatter(df_cust['Age'],df_cust['Spending Score (1-100)'],c=labels,s=10)
plt.scatter(centroids[:,0],centroids[:,1],c='red',marker='*',s=100)
```

# Agglomerative clustering - Bottom up approach of clustering

each sample point is taken as a cluster and then two shortest distances groups to form a cluster


```python
from sklearn import datasets
```

```python
Iris=datasets.load_iris()
print(Iris.feature_names)
print(Iris.target_names)
```

```python
df_Iris=pd.DataFrame(
                        {'Sepal length':Iris.data[:,0],
                         'Sepal width': Iris.data[:,1],
                         'Petal length':Iris.data[:,2],
                         'Petal width' :Iris.data[:,3],
                         'Breed':Iris.target}
                    )
```

```python
Iris_arr=np.array(df_Iris[['Sepal length','Petal length']])
x=df_Iris['Sepal length']
y=df_Iris['Petal length']
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agg_Iris=AgglomerativeClustering(n_clusters=3)
agg_Iris=agg_Iris.fit(Iris_arr)
labels=agg_Iris.labels_
plt.scatter(x,y,c=labels,s=20)
```

```python
import scipy.cluster.hierarchy as sch
dend = sch.dendrogram(sch.linkage(Iris_arr, method='average'))
plt.xlabel('Sepal length')
plt.ylabel('Petal length')
plt.figure(figsize=(100,25))
plt.show()
```
