```python
import pandas as pd
```

```python
df_cust=pd.read_csv(r"C:\Users\Priyadarshini\Downloads\Python datasets\customer_churn.csv")
```

```python
df_cust.head()
```

```python

df_cust['TotalCharges'] = pd.to_numeric(df_cust['TotalCharges'],errors='coerce')
df_cust.dropna(inplace=True)
df_cust.drop(columns='customerID',inplace=True)
```

```python
df_cust['gender'] = df_cust['gender'].str.lower().map({'male':0,'female':1})
```

```python
df_cust.head()
```

```python
df_cust['Partner'] = df_cust['Partner'].map({'Yes': 1, 'No': 0})
df_cust['Dependents'] = df_cust['Dependents'].map({'Yes': 1, 'No': 0})
df_cust['PhoneService']= df_cust['PhoneService'].map({'Yes': 1,'No': 0})
df_cust['MultipleLines']=df_cust['MultipleLines'].map({'No phone service':0, 'No': 1, "Yes": 2})
df_cust['InternetService']=df_cust['InternetService'].map({'DSL':0, 'Fiber optic':1, 'No':2})
df_cust['OnlineSecurity']=df_cust['OnlineSecurity'].map({'Yes':1, 'No':0, 'No internet service':2})
df_cust['OnlineBackup']=df_cust['OnlineBackup'].map({'Yes':1, 'No':0, 'No internet service':2})
df_cust['DeviceProtection']=df_cust['DeviceProtection'].map({'Yes':1, 'No':0, 'No internet service':2})
df_cust['TechSupport']=df_cust['TechSupport'].map({'Yes':1, 'No':0, 'No internet service':2})
df_cust['StreamingTV']=df_cust['StreamingTV'].map({'Yes':1, 'No':0, 'No internet service':2})
df_cust['StreamingMovies']=df_cust['StreamingMovies'].map({'Yes':1, 'No':0, 'No internet service':2})
df_cust['Contract']=df_cust['Contract'].map({'Month-to-month':0, 'One year':1, 'Two year':2})
df_cust['PaperlessBilling']=df_cust['PaperlessBilling'].map({'Yes':1, 'No':0})
df_cust['PaymentMethod']=df_cust['PaymentMethod'].map({'Electronic check':1, 'Mailed check':0, 'Bank transfer (automatic)':2, 'Credit card (automatic)':3})
df_cust['Churn']=df_cust['Churn'].map({'Yes':1,'No':0})
```

```python
df_cust.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=pd.DataFrame(df_cust.iloc[:,:-1])
y=pd.DataFrame(df_cust.iloc[:,-1])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc= RandomForestClassifier(n_estimators=500)
rfc.fit(x_train,y_train)
ypred=rfc.predict(x_test)
```

```python
from sklearn.metrics import *
```

```python
print("Accuracy Score",accuracy_score(ypred,y_test))
```

```python
importance=pd.Series(rfc.feature_importances_)
```

```python
df_imp=pd.DataFrame(columns=['Feature','Importance'])
```

```python
cols=list(df_cust.keys())
cols=cols[:-1]
```

```python
df_imp['Feature']=cols
df_imp['Importance']=importance
```

```python
df_imp
```

```python
df_imp.sort_values(by='Importance',ascending=True,ignore_index=True,inplace=True)
df_imp
```

```python
proba_valid=rfc.predict_proba(x_test)[:,1]
proba_valid
```

```python
proba_valid.shape
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,proba_valid)
```

```python
from sklearn.metrics import roc_curve
from matplotlib import pyplot
import numpy as np
```

```python
auc = roc_auc_score(np.array(y_test),proba_valid)
auc
```

```python
fpr,tpr,threshold = roc_curve(np.array(y_test),proba_valid)
```
