# Machine Learning

```python
import sklearn.metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import math
```

```python
x=pd.DataFrame(df_cust['tenure'])
y=pd.DataFrame(df_cust['MonthlyCharges'])
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=3,test_size=0.2)
print(x_train.shape)
```

```python
cust_LinReg=LinearRegression()
cust_LinReg.fit(x_train,y_train)
y_predict=cust_LinReg.predict(x_test)
MAE=sklearn.metrics.mean_absolute_error(y_test,y_predict)
MSE=sklearn.metrics.mean_squared_error(y_test,y_predict)
RMSE=math.sqrt(MSE)
print("MAE",MAE,"\nMSE",MSE,"\nRMSE",RMSE)
print("\nRegression intercept",cust_LinReg.intercept_)  #C
print("\nRegression Coeff",cust_LinReg.coef_)  #M
```

```python
x=pd.DataFrame(df_cust['MonthlyCharges'])
y=pd.DataFrame(df_cust['Churn'])
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.35)
```

```python
Log_reg=LogisticRegression()
Log_reg.fit(x_train,y_train)
y_pred=Log_reg.predict(x_test)
print(sklearn.metrics.confusion_matrix(y_test,y_pred))
print(sklearn.metrics.accuracy_score(y_test,y_pred))
```

```python
x=pd.DataFrame(df_cust['tenure'])
y=pd.DataFrame(df_cust['Churn'])
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.2)
```

```python
Dec_Tree=DecisionTreeClassifier()
Dec_Tree.fit(x_train,y_train)
y_pred=Dec_Tree.predict(x_test)
print(sklearn.metrics.confusion_matrix(y_test,y_pred))
print(sklearn.metrics.accuracy_score(y_test,y_pred))
```

```python
rand=RandomForestClassifier()
rand.fit(x_train,y_train)
y_pred=rand.predict(x_test)
print(sklearn.metrics.confusion_matrix(y_test,y_pred))
print(sklearn.metrics.accuracy_score(y_test,y_pred))
```
